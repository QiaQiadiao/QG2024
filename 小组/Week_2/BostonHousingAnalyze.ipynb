```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'  # 替换为你选择的字体
data = pd.read_csv(r"D:\学习\QG工作室\QG工作室2024训练营第二周\housing.csv")
print(data.describe())
```

```python
#画个波士顿城镇河景房的比例图
x=data.groupby('CHAS').size()
plt.pie(x,explode=[0.05,0.04],labels=['NotNearRiver','NearRiver'],autopct='%2.1f%%')
plt.title('波士顿城镇河景房的比例图')
plt.show()
```

```python
#画个波士顿城镇人均犯罪率
crime = data['CRIM']
crime_number,_=np.histogram(crime,bins=18,range=(0,91))
a = np.arange(0,90,5)
plt.bar(a+2.5,crime_number,edgecolor='black',width=5)
plt.xticks(range(0,91,5))
plt.yticks([5,100,200,300,400])
plt.title("波士顿城镇人均犯罪率分布表")
plt.xlabel('犯罪率/%')
plt.ylabel('城镇个数/个')
```
